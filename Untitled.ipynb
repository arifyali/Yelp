```python
import pandas as pd
```

```python
df = pd.read_csv('MedianValueSqFt_by_zip.csv', converters={'RegionID': lambda x: str(x)})
```

```python
df2 = pd.read_csv('ACS.csv', converters={'zip.code.tabulation.area.': lambda x: str(x)})
```

```python
df2.head()
```

```python
df2['RegionID'] = df2['zip.code.tabulation.area.']
```

```python
res_df = df.merge(df2, how='right', on='RegionID')
```

```python
res_df.reset_index(drop=True).head()
```

```python
res_df.count()
```

```python
res_df.to_csv('right_joined_census.csv')
```
